### First Attempt at Datasets Analysis

-This my first attempt at analysing the datasets that I collected for my individual project

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/arms-industry/defense-industry-data.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Company Name'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Companies by Number of Entries')
plt.xlabel('Company Name')
plt.ylabel('Number of Entries')
```

```python
df['Company Name'].unique()
```

```python
df['Volume'].hist(bins=30)
plt.title('Distribution of Volume')
plt.xlabel('Volume')
plt.ylabel('Frequency')
```

```python
df['Company Name'].head(10).describe()
df['Volume'].mean()
df['Volume'].max()
df['Volume'].min()
df['Volume'].std()
df['Volume'].median()
df.plot(x='Company Name', y='Volume', kind='scatter')
plt.title('Company Name vs Volume')
plt.xlabel('Company Name')
plt.ylabel('Volume')
plt.show()
```
